```python
import findspark
findspark.init()
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
from pyspark.sql.functions import *
import colorama
from colorama import *
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
# df=pd.read_csv("C:\\Users\\HP\\OneDrive\\Dataset.csv")
df=spark.read.csv("C:\\Users\\HP\\OneDrive\\Dataset.csv",inferSchema=True,header=True)
def plotting() :
     while True :
            print("1. Age Visualization With the help of dataset : ")
            print("2. Visualising BMI in the given dataset : ")
            print("3. Children Visualisation with the available dataset : ")
            print("4. Gender Visualization With the help of dataset : ")
            print("5. Charge Visualisation for the given dataset : ")
            print("6. Vizualisation Between sex and smoker in the given dataset : ")
            print("7. Vizualisation Between smokers and age in the given dataset : ")
            print("8. Smokers analysis on basis of gender from the given dataset : ")
            print("9. Analysing AGE with BMI,GENDER,SMOKER,CHARGES,CHILDREN : ")
            print("10. Analysing CHILDREN with AGE,GENDER,SMOKER,CHARGES,BMI : ")
            print("11. Analysing BMI with AGE,GENDER,SMOKER,CHARGES,CHILDREN : ")
            ch = int(input("Enter a number to view charting : "))
            if ch == 1 :
                print("Age Visualization With the help of dataset")
                plt.figure()
                plt.figure(figsize=(12,10))
                sn.countplot(x='age',data=df,palette='rainbow')
                break
            elif ch == 2 :
                print("Visualising BMI in the given dataset")
                plt.figure(figsize=(12,10))
                plt.figure()
                sn.histplot(df.bmi,color='violet',kde=True, stat="density", linewidth=0)
                break
            elif ch==3:
                print("Children Visualisation with the available dataset")
                plt.figure()
                plt.figure(figsize=(12,10))
                sn.countplot(x='children',data=df)
                break
            elif ch==4:
                print("Gender Visualization With the help of dataset")
                plt.figure()
                df.sex.value_counts().plot(kind='bar',color='indigo')
                break
            elif ch==5:
                print("Charge Visualisation for the given dataset")
                plt.figure()
                df.charges.plot(kind='bar',color='black')
                break
            elif ch==6:
                print("Vizualisation Between sex and smoker in the given dataset")
                plt.figure()
                plt.figure(figsize=(12,10))
                plt.figure()
                sn.countplot(x='sex',hue='smoker',data=df)
                break
            elif ch==7:
                print("Vizualisation Between smokers and age in the given dataset")
                plt.figure()
                sn.countplot(x='age',hue='smoker',data=df)
                break
            elif ch==8:
                print("Smokers analysis on basis of gender from the given dataset : ")
                plt.figure()
                df.groupby(['sex','smoker']).mean().plot(kind='bar',color='black')
                break
            elif ch==9:
                print("Analysing AGE with BMI,GENDER,SMOKER,CHARGES,CHILDREN")
                f,ax=plt.subplots(nrows=6,ncols=1,figsize=(15,25))
                for i,col in enumerate (df.drop('age',axis=1).columns):
                    sn.lineplot(x=df['age'],y=df[col],ax=ax[i],color='green')
                break
            elif ch==10:
                print("Analysing CHILDREN with AGE,GENDER,SMOKER,CHARGES,BMI")
                f,ax=plt.subplots(nrows=6,ncols=1,figsize=(15,25))
                for i,col in enumerate (df.drop('children',axis=1).columns):
                    sn.boxplot(x=df['children'],y=df[col],ax=ax[i])
                break
            elif ch==11:
                print("Analysing BMI with AGE,GENDER,SMOKER,CHARGES,CHILDREN")
                f,ax=plt.subplots(nrows=6,ncols=1,figsize=(15,25))
                for i,col in enumerate (df.drop('bmi',axis=1).columns):
                    sn.lineplot(x=df['bmi'],y=df[col],ax=ax[i])
                break
            else:
                print("provide the options mentioned above : ")
plotting()
```

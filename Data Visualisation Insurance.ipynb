```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
df=pd.read_csv("C:\\Users\\HP\\OneDrive\\Dataset.csv")
```

```python
# Age Visualisation In Dataset
print("Age Visualization With the help of dataset")
plt.figure()
plt.figure(figsize=(12,10))
sn.countplot(x='age',data=df,palette='rainbow')
```

```python
print("Visualising BMI in the given dataset")
plt.figure(figsize=(12,10))
plt.figure()
sn.histplot(df.bmi,color='violet',kde=True, stat="density", linewidth=0)
```

```python
#children dataset analysis
print("Children Visualisation with the available dataset")
plt.figure()
plt.figure(figsize=(12,10))
sn.countplot(x='children',data=df)
```

```python
#gender Visualisation
print("Gender Visualization With the help of dataset")
plt.figure()
df.sex.value_counts().plot(kind='bar',color='indigo')
```

```python
# Charge Visualisation for the given dataset
print("Charge Visualisation for the given dataset")
plt.figure()
df.charges.plot(kind='bar',color='black')
```

```python
print("Vizualisation Between sex and smoker in the given dataset")
plt.figure()
plt.figure(figsize=(12,10))
plt.figure()
sn.countplot(x='sex',hue='smoker',data=df)
```

```python
print("Vizualisation Between smokers and age in the given dataset")
plt.figure()
sn.countplot(x='age',hue='smoker',data=df)
```

```python
print("Smokers analysis on basis of gender from the given dataset : ")
plt.figure()
df.groupby(['sex','smoker']).mean().plot(kind='bar',color='black')
```

```python
print("Analysing AGE with BMI,GENDER,SMOKER,CHARGES,CHILDREN")
f,ax=plt.subplots(nrows=6,ncols=1,figsize=(15,25))
for i,col in enumerate (df.drop('age',axis=1).columns):
    sn.lineplot(x=df['age'],y=df[col],ax=ax[i],color='green')
```

```python
print("Analysing CHILDREN with AGE,GENDER,SMOKER,CHARGES,BMI")
f,ax=plt.subplots(nrows=6,ncols=1,figsize=(15,25))
for i,col in enumerate (df.drop('children',axis=1).columns):
    sn.boxplot(x=df['children'],y=df[col],ax=ax[i])
```

```python
print("Analysing BMI with AGE,GENDER,SMOKER,CHARGES,CHILDREN")
f,ax=plt.subplots(nrows=6,ncols=1,figsize=(15,25))
for i,col in enumerate (df.drop('bmi',axis=1).columns):
    sn.lineplot(x=df['bmi'],y=df[col],ax=ax[i])
```
